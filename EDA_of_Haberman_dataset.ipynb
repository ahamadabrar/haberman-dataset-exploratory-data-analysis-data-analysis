<h3>CDF</h3>

```python
# Plots of CDF of age for survival status class 1 & 2.

hist1, bin_edges1 =np.histogram(survival_status_class1['age'], bins= 10, density= True)
# print("hist_size= {}\nbin_edges_size= {}".format(hist.size,bin_edges.size))
"""
Note::
 hist1 is having 10 values
 bin_edges1 is having (bin+1=10+1=11) values
 so when we plot the pdf or cdf, we'll take (bin_edges.size-1) value
"""
#finding pdf1 using hist1 numpy array
#Note:: "hist1/(sum(hist1))" returns an array that will store in "pdf"
pdf1 = hist1/(sum(hist1))
#calculating cdf using "cumsum()"
cdf1 = np.cumsum(pdf1)

hist2, bin_edges2 = np.histogram(survival_status_class2['age'], bins= 10, density= True)
pdf2 = hist2/(sum(hist2))
cdf2 = np.cumsum(pdf2)


fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(bin_edges1[1:], pdf1, label = 'PDF of servival_status_1')
axs[0].plot(bin_edges1[1:], cdf1, label = 'CDF of servival_status_1')
axs[1].plot(bin_edges2[1:], pdf2, label = 'PDF of servival_status_2')
axs[1].plot(bin_edges2[1:], cdf2, label = 'CDF of servival_status_2')

axs[0].legend()
axs[1].legend()
# fig.suptitle('Patients who servived 5yrs or longer                Patients who died within 5yrs')
fig.supxlabel('patients age')
fig.supylabel('% of total distribution')
```

***Observations***

1. 20% of total patients who servived 5 yrs or longer had age <= 42yr (approx)
  
  - total patients who servived 5 yrs or longer = 225 <br>so 20% of 225 = 45 patients <br> i.e. 45 patients out of 225  who servived 5 yrs or longer had age <= 42yr (approx)
2. approx 70% of total patients who servived 5 yrs or longer had age <= 60yr (approx)
3. approx 8% of total patients who died within 5 yrs had age <= 42yr (approx)
 - total patients who died within 5yrs = 81 <br>so 8% of 81 = 6(approx) patients <br> i.e. 6 patients out of 81 who died within 5 yrs had age <= 42yr (approx)
4. approx 65% of total patients who died within 5 yrs had age <= 60yr (approx)
5. approx 90% of total patients who died within 5 yrs had age <= 70yr (approx)

```python
# Plots of CDF of operation_Year for survival status class 1 & 2.


hist1, bin_edges1 =np.histogram(survival_status_class1['operation_Year'], bins= 10, density= True)
# print("hist_size= {}\nbin_edges_size= {}".format(hist.size,bin_edges.size))
"""
Note::
 hist is having 10 values
 bin_edges is having (bin+1=10+1=11) values
 so when we plot the pdf or cdf, we'll take (bin_edges.size-1) value
"""
#finding pdf using hist numpy array
#Note:: "hist/(sum(hist))" returns an array that will store in "pdf"
pdf1 = hist1/(sum(hist1))
#calculating cdf using "cumsum()"
cdf1 = np.cumsum(pdf1)


hist2, bin_edges2 = np.histogram(survival_status_class2['operation_Year'], bins= 10, density= True)
pdf2 = hist2/(sum(hist2))
cdf2 = np.cumsum(pdf2)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(bin_edges1[1:], pdf1, label = 'PDF of servival_status_1')
axs[0].plot(bin_edges1[1:], cdf1, label = 'CDF of servival_status_1')
axs[1].plot(bin_edges2[1:], pdf2, label = 'PDF of servival_status_2')
axs[1].plot(bin_edges2[1:], cdf2, label = 'CDF of servival_status_2')

axs[0].legend()
axs[1].legend()

fig.supxlabel('operation_Year')
fig.supylabel('% of total distribution')
```

```python
# Plots of CDF of pos_axillary_nodes for survival status class 1 & 2.


hist1, bin_edges1 =np.histogram(survival_status_class1['pos_axillary_nodes'], bins= 10, density= True)
# print("hist_size= {}\nbin_edges_size= {}".format(hist.size,bin_edges.size))
"""
Note::
 hist is having 10 values
 bin_edges is having (bin+1=10+1=11) values
 so when we plot the pdf or cdf, we'll take (bin_edges.size-1) value
"""
#finding pdf using hist numpy array
#Note:: "hist/(sum(hist))" returns an array that will store in "pdf"
pdf1 = hist1/(sum(hist1))
#calculating cdf using "cumsum()"
cdf1 = np.cumsum(pdf1)

hist2, bin_edges2 = np.histogram(survival_status_class2['pos_axillary_nodes'], bins= 10, density= True)
pdf2 = hist2/(sum(hist2))
cdf2 = np.cumsum(pdf2)

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].plot(bin_edges1[1:], pdf1, label = 'PDF of servival_status_1')
axs[0].plot(bin_edges1[1:], cdf1, label = 'CDF of servival_status_1')
axs[1].plot(bin_edges2[1:], pdf2, label = 'PDF of servival_status_2')
axs[1].plot(bin_edges2[1:], cdf2, label = 'CDF of servival_status_2')

axs[0].legend()
axs[1].legend()

fig.supxlabel('pos_axillary_nodes')
fig.supylabel('% of total distribution')
```
